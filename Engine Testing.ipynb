```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
path='/content/drive/MyDrive/DAL/automobileEDA.csv'
df_automobile = pd.read_csv(path)
df_automobile.head()
```

```python
df_automobile.shape
```

```python
df_automobile.info()
```

```python
df_data = df_automobile.replace('?',np.NAN)
df_data.isnull().sum()
```

```python
df_automobile['stroke'].fillna('np.nan', inplace= True)
```

```python
df_automobile['horsepower-binned'].fillna('np.nan', inplace= True)
```

```python
df_automobile.isnull().sum()
```

```python
df_automobile.head(10)
```

```python
df_automobile.describe()
```

```python
plt.figure(figsize=(10,8))
df_automobile[['engine-size','peak-rpm','curb-weight','horsepower','price']].hist(figsize=(10,10),bins=6,color='Y')
plt.figure(figsize=(10,8))
plt.tight_layout()
plt.show()
```

```python
plt.figure(1)
plt.subplot(221)
df_automobile['engine-type'].value_counts(normalize=True).plot(figsize=(10,8),kind='bar',color='red')
plt.title("Number of Engine Type frequency diagram")
plt.ylabel('Number of Engine Type')
plt.xlabel('engine-type');


plt.subplot(222)
df_automobile['num-of-doors'].value_counts(normalize=True).plot(figsize=(10,8),kind='bar',color='green')
plt.title("Number of Door frequency diagram")
plt.ylabel('Number of Doors')
plt.xlabel('num-of-doors');



plt.subplot(224)
df_automobile['body-style'].value_counts(normalize=True).plot(figsize=(10,8),kind='bar',color='orange')
plt.title("Number of Body Style frequency diagram")
plt.ylabel('Number of vehicles')
plt.xlabel('body-style');
plt.tight_layout()
plt.show()
```
